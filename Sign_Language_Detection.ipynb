```python
import os
import hashlib
from google.colab import drive

def file_hash(filepath):
    with open(filepath, 'rb') as f:
        return hashlib.md5(f.read()).hexdigest()

def find_duplicates(directory):
    hash_dict = {}
    duplicates = []

    for root, dirs, files in os.walk(directory):
        for file in files:
            filepath = os.path.join(root, file)
            file_hash_value = file_hash(filepath)

            if file_hash_value in hash_dict:
                duplicates.append((filepath, hash_dict[file_hash_value]))
            else:
                hash_dict[file_hash_value] = filepath

    return duplicates

# Mount Google Drive
drive.mount('/content/drive')

# Specify the directory where the files are located
directory_path = '/content/drive/My Drive/Sign Language for Alphabets/a'

# Find and display duplicate files
duplicate_files = find_duplicates(directory_path)

if duplicate_files:
    print("Duplicate Files:")
    for duplicate_pair in duplicate_files:
        print(f'Duplicate Pair: {duplicate_pair}')
else:
    print("No duplicate files found.")
```

```python
import os
import hashlib
from google.colab import drive

def file_hash(filepath):
    with open(filepath, 'rb') as f:
        return hashlib.md5(f.read()).hexdigest()

def find_duplicates_and_delete(directory):
    hash_dict = {}
    duplicates = []

    for root, dirs, files in os.walk(directory):
        for file in files:
            filepath = os.path.join(root, file)
            file_hash_value = file_hash(filepath)

            if file_hash_value in hash_dict:
                duplicates.append((filepath, hash_dict[file_hash_value]))
            else:
                hash_dict[file_hash_value] = filepath

    # Delete duplicate files
    for duplicate_pair in duplicates:
        print(f'Deleting Duplicate Pair: {duplicate_pair}')
        os.remove(duplicate_pair[0])

# Mount Google Drive
drive.mount('/content/drive')

# Specify the directory where the files are located
directory_path = '/content/drive/My Drive/Sign Language for Alphabets/a'

# Find and delete duplicate files
find_duplicates_and_delete(directory_path)
```
